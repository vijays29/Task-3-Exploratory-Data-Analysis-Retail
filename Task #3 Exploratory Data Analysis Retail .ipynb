# Data visualization

```python
plt.figure(figsize=(9,8))
sns.heatmap(df.corr(),annot=True)
plt.show()
```

```python
sns.pairplot(df)
plt.show()
```

```python
x=df.groupby(by="Category")[['Profit']].sum()
b=x.sort_values(by=['Profit'],ascending=False)
b.plot(kind='bar')
plt.show()
```

```python
plt.figure(figsize=(9,7))
sns.barplot(x='Region',y='Profit', hue='Category',data=df)
plt.show()
```

```python
plt.figure(figsize=(9,6))
sns.barplot(x='Region',y='Sales', hue='Category',data=df)
plt.show()
```

```python
plt.figure(figsize=(3,3))
df.groupby('Region')['Profit'].sum().plot(kind='pie',autopct='%1.1f%%')
plt.show()
```

```python
category=df.groupby(by="Sub-Category")[['Profit','Sales']].sum()
profit=category.sort_values(by=['Profit'],ascending=True).plot(kind='bar',figsize=(10,8))
```
