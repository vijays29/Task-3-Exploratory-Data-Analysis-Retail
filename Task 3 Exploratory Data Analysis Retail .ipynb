# Data Visualization Using Pairplot

```python
plt.figure(figsize=(8,5))
sns.kdeplot(df['Sales'],color='red',label='Sales',shade=True)
sns.kdeplot(df['Profit'],color='Blue',label='Profit',shade=True)
plt.xlim([-100,1000])
plt.legend()
```

# Profit Discount and Sale

```python
df['Segment'].value_counts()
```

```python
df_segment= df.groupby(['Segment'])[['Sales', 'Discount', 'Profit']].mean()
df_segment
```

```python
x=df.groupby(by="Category")[['Profit']].sum()
b=x.sort_values(by=['Profit'],ascending=False)
b.plot(kind='bar')
plt.show()
```

```python
plt.figure(figsize=(9,5))
sns.barplot(x='Region',y='Profit', hue='Category',data=df)
plt.show()
```

```python
plt.figure(figsize=(6,8))
sns.barplot(x='Region',y='Sales', hue='Category',data=df)
plt.show()
```

```python
plt.figure(figsize=(5,7))

df.groupby('Region')['Profit'].sum().plot(kind='pie',autopct='%1.1f%%')
plt.show()
```

```python
subcategory=df.groupby(by="Sub-Category")[['Profit','Sales']].sum()
subprofit=subcategory.sort_values(by=['Profit'],ascending=True).plot(kind='bar',figsize=(10,8))
```
